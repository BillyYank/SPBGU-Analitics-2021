Яндекс places

```python
places_url = 'https://search-maps.yandex.ru/v1/'
result = json.loads(requests.get(places_url, params={
    'apikey': placeskey,
    'text': 'Шоколадницы Москва',
    'lang': 'ru',
    'results': 500
}).content)
```

```python
result
```

```python
phones = []
addreses = []

for el in result["features"]:
    try:
        phones.append(el['properties']['CompanyMetaData']['Phones'][0]['formatted'])
        addreses.append(el['properties']['CompanyMetaData']['address'])
    except Exception as e:
        print(e)
```

```python
choko_data = pd.DataFrame({
    'Телефон': phones,
    'Адрес': addreses
})
```

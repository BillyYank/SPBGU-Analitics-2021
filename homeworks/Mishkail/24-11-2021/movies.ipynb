```python
# создадим объект, в котором толлько нужные столбцы из метадата
meta_1 = meta[['id','title', 'budget', 'revenue', 'vote_average', 'vote_count', 'popularity']] 
```

```python
meta_1.head()
```

```python
meta_1 = meta_1[meta_1['budget'].apply(lambda x: x.isdigit())]
```

```python
# преобразуем столбцы в числовые значения
meta_1['budget'] = pd.to_numeric(meta_1['budget'])
meta_1['popularity'] = pd.to_numeric(meta_1['popularity'])
```

```python
meta_1.info()
```

```python
meta_1.sort_values('budget', ascending='True') # мешают фильмы с нулевым бюджетом
```

```python
# убираем фильмы с нулевым бюджетом
meta_1 = meta_1[meta_1['budget'] != 0]
```

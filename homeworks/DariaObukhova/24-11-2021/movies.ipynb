```python
import pandas as pd
import os

%matplotlib inline
```

```python
movies = pd.read_csv('data/movies_metadata.csv')
movies.head()
```

```python
movies = movies[movies['budget'].apply(lambda x: x.isdigit())]
movies.columns
```

```python
movies['budget'] = movies['budget'].apply(float)
```

```python
movies.dtypes
```

```python
movies.shape
```

Самый дорогой и самый дешевый фильмы

```python
MaxMovieBudget = movies['budget'].max() 
MaxMovieBudget
```

```python
MaxMovieBudget = movies[movies['budget']>=MaxMovieBudget]
MaxMovieBudget = MaxMovieBudget[['budget','original_title']]
MaxMovieBudget
```

```python
MinMovieBudget = movies[movies['budget'] != 0]
MinMovieBudget = movies['budget'].min()
MinMovieBudget
```

```python
MinMovieBudget = movies[movies['budget']>=MinMovieBudget]
MinMovieBudget = MinMovieBudget[['budget','original_title']]
MinMovieBudget
```

Топ 100 самых лучших и самых худших фильмов

```python
movies = movies[['original_title','vote_average']]
```

```python
movies
```

```python
BestMovies = movies.sort_values(by = 'vote_average', ascending=False)
```

```python
BestMovies.head(100)[['original_title','vote_average']]
```

```python
WorstMovies = movies.sort_values(by = 'vote_average', ascending=True)
```

```python
WorstMovies = WorstMovies[WorstMovies['vote_average'] != 0]
```

```python
WorstMovies.head(100)[['original_title','vote_average']]
```

Ужасные, но популярные фильмы

```python
movies = movies[['original_title','vote_average', 'popularity']]
```

```python
movies
```

```python
movies = movies[['original_title','vote_average','popularity']]
movies = movies.sort_values(['vote_average', 'popularity'], ascending=[True, False])
movies.head(10)
```

Выше 10 фильмов популярных, но с низким рейтингом (скорее всего, по рейтингу нет данных)

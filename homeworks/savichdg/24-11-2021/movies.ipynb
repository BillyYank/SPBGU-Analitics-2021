# 4. Cамые дорогие и самые дешевые жанры в плане производства. Самые популярные и самые непопулярные жанры

```python
genres_info = movies[['id', 'genres']]


#genres_info.columns = ['movieId', 'genres']
#genres_info = genres_info.rename(columns = {'id':'movieId'})

# genres_info.head(25)
# genres_info['genres'][2]
# "[{'id': 10749, 'name': 'Romance'}, {'id': 35, 'name': 'Comedy'}]"
#type(genres_info['genres'][2])
# str
#len

allGenresList = {} # Глобальный словарь с названиями жанров

def genresFromJSON(x):
    # Преобразование данных (первичное, в список) из JSON для столбца датайрейма и заполнение глобального списка жанров
    x = json.loads(x.replace("'", '"'))
    # Обратное - json.dumps() Python в JSON
    for listval in x:
        if listval['name'] not in allGenresList:
            allGenresList[listval['id']] = listval['name']            
    return (x)

genres_info['genres'] = genres_info['genres'].apply(lambda x: genresFromJSON(x))

def genresFromListToArray(x):
    resArr = []
    for listval in x:
        resArr.append(listval['id']) 
    return (resArr)
genres_info['genres'] = genres_info['genres'].apply(lambda x: genresFromListToArray(x))
genres_info
```

```python

movies_for_budget = movies_for_budget.merge(genres_info, on='id')

```

````python
sum_diag,i,j=0,0,0
A = np.random.normal(size=(10, 5))
B = np.random.normal(size=(5, 10))
total = A.dot(B)
while i<10:
  j=i
  sum_diag=sum_diag+total[i,j]
  i+=1
sum_diag
````

````python
sum_diag_1=np.trace(total)
sum_diag_1
````

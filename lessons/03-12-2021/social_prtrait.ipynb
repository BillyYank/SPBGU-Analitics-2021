```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('Новая форма (Ответы) (1).xlsx', sheet_name='Ответы на форму (1)')
```

```python
data.drop(['Отметка времени', '2. Город проживания'], axis=1, inplace=True)
```

```python
data = data[data.columns[0:7]]
```

```python
(data[data.columns[3]] - data[data.columns[4]]).value_counts()
```

```python
X = data.drop(data.columns[3:5], axis=1)
y = data[data.columns[4]] >= 3
```

```python
X_num = pd.get_dummies(X)
```

```python
X_num
```

```python
y.value_counts()
```

```python
X_num.fillna(-1, inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
dt = DecisionTreeClassifier(max_depth=4)
```

```python
dt.fit(X_num, y, )
```

```python
accuracy_score(y, dt.predict(X_num)) 
```

```python
recall_score(y, dt.predict(X_num))
```

```python
precision_score(y, dt.predict(X_num))
```

```python
from sklearn import tree
import pydotplus

dot_data=tree.export_graphviz(dt,filled=True,rounded=True)
graph=pydotplus.graph_from_dot_data(dot_data)
graph.write_png('tree.png')    
plt.imshow(plt.imread('tree.png'))
```
